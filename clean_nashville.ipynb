```python
import pandas as pd
```

1. Load the dataset

- Use pandas to read the CSV file.

- Display the first 5 rows of the dataset.

- Use .info() and .describe() to profile the data.

```python
housing_data = pd.read_csv("Nashville_housing_data.csv")
housing_data[:5]
```

```python
housing_data.info()
housing_data.describe()
```

2. Standardize Headings

Change all column headings to be all lower-case lettering. If a column heading has two or more words, then connect each word using "_".  For example, "Column Heading" would become "column_heading".

```python
housing_data.columns = housing_data.columns.str.lower().str.replace(" ", "_")
```

3. Eliminate Unneeded Columns

Eliminate columns we don't need: unnamed_0.1 and unnamed_0.

```python
housing_data = housing_data.drop(columns=["unnamed:_0.1", "unnamed:_0"])
```

4. Handle missing values
Drop all rows where the sale_price column is missing.

Fill missing values in the Bedrooms column with the median number of bedrooms.

```python
housing_data = housing_data.dropna(subset=['sale_price'])
housing_data['bedrooms'] = housing_data['bedrooms'].fillna(housing_data['bedrooms'].median())
```

5. Fix column data types
Convert the sale_date column to datetime format using pd.to_datetime().

```python
housing_data['sale_date'] = pd.to_datetime(housing_data['sale_date'])
```
