Clean up column names
- Start with "Suite/ Condo   #" since it has the most custom format ending with a hash
- Make columns lowercase and remove non-alphanumeric characters
- Replace spaces with underscores

```python
#df = df.rename(columns={"Suite/ Condo   #": "suite_condo_number"})
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
```

Drop first two unused columns

```python
df = df.drop(columns=["unnamed:_0.1", "unnamed:_0"])
```

Remove rows with missing sale price

```python
df = df.dropna(subset=['sale_price'])
```

Set rows with missing bedroom count to the median number of bedrooms

df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())

Set data type of sale_date to date/time

```python
df['sale_date'] = pd.to_datetime(df['sale_date'])
```

Set all sold_as_vacant values to 'Yes' or 'No' cleaning up inconsistent values like 'Y' or 'N'

```python
df['sold_as_vacant'] = df['sold_as_vacant'].str.strip().str.upper().str[0].map({'Y': 'Yes', 'N': 'No'})
```

Save results to csv

```python
df.to_csv("nashville_housing_cleaned.csv", index=False)
```
