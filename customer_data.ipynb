Import libraries

```python
import pandas as pd
```

Read in data

```python
customers_df = pd.read_csv('customers.csv')
orders_df = pd.read_csv('orders.csv')
products_df = pd.read_csv('products.csv')
```

Merged data

```python
merged_order_product_df = pd.merge(orders_df, products_df, on='product_id', how='left')
cd = pd.merge(merged_order_product_df, customers_df, on='customer_id', how='left')
cd.to_csv('master_data.csv', index=False)
```

```python
cd['order_date'] = pd.to_datetime(cd['order_date'])
```

```python
cd['revenue'] = cd['quantity'] * cd['price']
cd['order_month'] = cd['order_date'].dt.month_name()
```

```python
cd.columns = cd.columns.str.lower()
print(cd.head(5))
```

Sales Metrics
- Compute the total revenue across all orders.

- List the top 5 products by total revenue, sorted in descending order.

- List the top 5 products by total quantity sold, sorted in descending order.

- Calculate the average order value (aov) as:
aov = total revenue / total number of orders

-- Round aov to 2 decimal places.

```python
# Total revenue
cd['revenue'].sum()
```

```python
# Top 5 products by revenue
top_products = cd.groupby('product_name')['revenue'].sum().sort_values(ascending=False).head(5)
print(top_products)
```

```python
# Top 5 products by total quantity sold in descending order
top_products_quantity = cd.groupby('product_name')['quantity'].sum().sort_values(ascending=False).head(5)
print(top_products_quantity)
```

```python
# Average order value rounded to 2 decimal places. aov = total revenue / number of orders
aov = (cd['revenue'].sum() / cd['order_id'].nunique()).round(2)
print(aov)
```

Customer Metrics

- Count the total number of unique customers (unique_customers).

- Group the data by region and calculate:

- Total number of orders per region (orders_by_region)

- Total revenue per region (revenue_by_region)

- Identify the top 5 customers (top5_customers) by total spend (revenue), sorted in descending order.

```python
# Count total number of unique customers
cd['customer_id'].nunique()
```

```python
# Total number of order per region
cd.groupby('region')['order_id'].count()
```

```python
# Total revenue per region
cd.groupby('region')['revenue'].sum()
```

```python
# Identify the top 5 customers (top5_customers) by total spend (revenue), sorted in descending order.
top5_customers = cd.groupby('customer_name')['revenue'].sum().sort_values(ascending=False).head(5)
print(top5_customers)
```
